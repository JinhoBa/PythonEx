```python
import numpy as np
import pandas as pd
```

```python
exam = pd.read_csv('exam.csv')
exam.head()
```

```python
exam.groupby('nclass', as_index = False)\
    .agg( mean_math = ('math', 'mean'),
          mean_english = ('english', 'mean'),
          max_science = ('science', 'max'))
```

```python
exam.groupby('nclass').mean()
```

```python
exam.groupby('nclass')\
    .agg(mean_math = ('math', 'mean'),
         sum_math = ('math', 'sum'),
         median_math = ('math', 'median'),
         n = ('math','count'))
```

```python
mpg = pd.read_csv('mpg.csv')
mpg.head()
```

```python
mpg.groupby(['manufacturer', 'drv']).agg(mean_cty = ('cty','mean'),
                                         mean_hwy = ('hwy', 'mean'))\
    .sort_values('mean_cty', ascending =False).head(5)
```

```python
mpg.query('manufacturer == "audi"')\
    .groupby(['drv'])\
    .agg(n = ('drv', 'count'))
```

```python
mpg.value_counts('drv').to_frame('count').query('count > 100')
```

```python
mpg.head(5)
```

```python
mpg.query('category == "suv"')\
    .assign(total = (mpg['hwy']+mpg['cty']) / 2)\
    .groupby('manufacturer')\
    .agg(mean_tot = ('total', 'mean'))\
    .sort_values('mean_tot', ascending=False)\
    .head(5)
```

```python
mpg.groupby('category').agg(mean_cty=('cty','mean'))\
    .sort_values('mean_cty',ascending=False)
```

```python
mpg.groupby('manufacturer').agg(count_compact = ('category', 'count'))\
    .sort_values('count_compact', ascending=False)
```

```python
midwest = pd.read_csv('midwest.csv')
midwest.head(5)
```

```python
midwest['ratio'] = (midwest['poptotal'] - midwest['popadults']) / midwest['poptotal']*100
midwest[['county', 'state', 'ratio']].head(5)
```

```python
midwest.sort_values('ratio')[['state','county','ratio']].head(5)
```

```python
midwest['grade'] = np.where(midwest['ratio'] >= 40, 'large',
                    np.where(midwest['ratio'] >= 30, 'middle', 'small' ))
midwest[['grade','county','ratio']].head(5)
```

```python
midwest.groupby('grade').agg(count_grade = ('grade', 'count'))
```
